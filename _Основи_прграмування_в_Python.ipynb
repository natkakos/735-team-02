##Podmienky

–î–æ—Ä—ñ–≤–Ω—é—î: `a == b`

–ù–µ –¥–æ—Ä—ñ–≤–Ω—é—î: `a != b`

–ú–µ–Ω—à–µ –Ω—ñ–∂: `a < b`

–ú–µ–Ω—à–µ –∞–±–æ –¥–æ—Ä—ñ–≤–Ω—é—î: `a <= b`

```
print("–ó–∞–¥–∞–π—Ç–µ –ø–µ—Ä—à–µ —á–∏—Å–ª–æ")
a = int(input())
print("–ó–∞–¥–∞–π—Ç–µ –¥—Ä—É–≥–µ —á–∏—Å–ª–æ")
b = int(input())
print("–ó–∞–¥–∞–π—Ç–µ –æ–ø–µ—Ä–∞—Ü—ñ—é")
operacia = input()
if operacia == "*":
  vysledok = a*b
elif operacia == "/":
  vysledok = a/b
elif operacia == "+":
  vysledok = a+b
elif operacia == "-":
  vysledok = a-b
else:
  vysledok = "—Å—Ç–∞–ª–∞—Å—å –ø–æ–º–∏–ª–∫–∞"

print(vysledok)
```
